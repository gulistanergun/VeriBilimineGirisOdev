```python
import numpy as np
import pandas as pd
```

```python
"""
	1. Numpy ile Matris İşlemleri
- 5x5 boyutunda rastgele (0-100 arasında) tam sayılardan oluşan bir matris oluşturun.
- Bu matrisin:
  • Ortalama, standart sapma ve varyansını hesaplayın.
  • En büyük ve en küçük değerlerini bulun.
  • Köşegenindeki elemanların toplamını bulun.
"""
matris = np.random.randint(0,100,25).reshape(5, 5)
matris
```

```python
ortalama = np.mean(matris)
print(f'matris ortalaması: {ortalama:.2f}')
std_sapma=np.std(matris)
print(f'matris standart sapması: {std_sapma:.2f}')
varyans=np.var(matris)
print(f'matris varyans değeri: {varyans:.2f}')
en_kucuk= np.min(matris)
print(f'matris en küçük değeri: {en_kucuk}')
en_buyuk= np.max(matris)
print(f'matris en büyük değeri: {en_buyuk}')

print('\n')

kosegen_toplami = np.trace(matris)
print(f'matris köşegenindeki elemanların toplamı: {kosegen_toplami}')

kosegen_elemanlari = np.diag(matris)
print(f'matris köşegenindeki elemanlar: {kosegen_elemanlari}')

#kosegen_elemanlari_toplami = kosegen_elemanlari.sum()
#print(f'matris köşegenindeki elemanların toplamı: {kosegen_elemanlari_toplami}')
```

```python
#2.bölüm
"""
2.	2. Numpy ile Veri Simülasyonu
- 1000 adet öğrencinin sınav puanlarını (0-100 arası, normal dağılımdan gelen) simüle edin.
- Ortalama, medyan ve standart sapmayı hesaplayın.
- 50’den düşük alan kaç öğrenci olduğunu bulun.

"""
puanlar = np.random.normal(loc=70, scale=15, size=1000).astype(int)
kirpilmis_puanlar = np.clip(puanlar, 0, 100)
print(kirpilmis_puanlar)
print("\n")

#ortalam medyan ve standart sapamyı hesapla
print("Ortalama:", np.mean(kirpilmis_puanlar))
print("Medyan:", np.median(kirpilmis_puanlar))
print("Standart Sapma:", f"{np.std(kirpilmis_puanlar):.2f}")


print("\n")
# 50’den düşük alan kaç öğrenci var
print("50’den düşük alan öğrenci sayısı:", np.sum(puanlar < 50))


```

```python
"""
pandas kullanılarak dataframe kullanımı
3.	1. Veri setini Pandas DataFrame olarak oluşturun.
4.	2. Her ders için ortalama puanı bulun.
5.	3. En yüksek matematik notunu alan öğrenciyi bulun.
6.	4. Her öğrencinin not ortalamasını hesaplayan yeni bir sütun ekleyin.
7.	5. Bölümlere göre gruplayarak her bölümün ortalama başarılarını hesaplayın.
8.	6. Ortalaması 70’in üzerinde olan öğrencileri filtreleyin.
"""
import pandas as pd
data = {"Öğrenci": ["Ali", "Ayşe", "Mehmet", "Zeynep", "Ahmet"],
    "Yaş": [20, 21, 19, 22, 20],
    "Bölüm": ["Bilgisayar", "Fizik", "Kimya", "Bilgisayar", "Fizik"],
    "Matematik": [70, 60, 80, 90, 55],
    "Fizik": [65, 75, 70, 85, 60],
    "Kimya": [80, 85, 65, 95, 70]
}
df = pd.DataFrame(data)
df



```

```python
#ders ortlamalrını heseplama
ders_ort = df[['Matematik', 'Fizik', 'Kimya']].mean()
for ders, ortalama in ders_ort.items():
    print(f"{ders} Ortalaması: {ortalama}")

```

```python
# matemetik için en yüksek paunı bul
max_matematik = df['Matematik'].idxmax()
max_mat_puan = df.loc[max_matematik]
max_mat_puan

```

```python
#Her öğrencinin not ortalamasını hesaplayan yeni bir sütun ekleyin.
df['Not Ortalaması'] = df[['Matematik','Fizik','Kimya']].mean(axis=1)
df
```

```python
#5. Bölümlere göre gruplayarak her bölümün ortalama başarılarını hesaplayın.
bolum_ort = df.groupby('Bölüm')[['Matematik','Fizik','Kimya','Not Ortalaması']].mean()
bolum_ort.round(2)

```

```python
#6. Ortalaması 70’in üzerinde olan öğrencileri filtreleyin.
basari = df[df['Not Ortalaması'] > 70]
basari
```

```python
#Matplotlib veya Seaborn kullanarak:
# Her dersin dağılımını histogram ile gösterin.
# Bölümlere göre ortalamaları bar grafikte görselleştirin.

# her dersin dağılımını histogram ile göstermek
import matplotlib.pyplot as plt
import seaborn as sns

dersler = ["Matematik", "Fizik", "Kimya"]

plt.figure(figsize=(8, 3))

for i, ders in enumerate(dersler):
    plt.subplot(1, 3, i+1)  # 1 satır, 3 sütun, i+1'inci subplot
    sns.histplot(df[ders], bins=5, kde=True, color="green")
    plt.title(f"{ders} Dağılımı")
    plt.xlabel("Not")
    plt.ylabel("Öğrenci Sayısı")

plt.tight_layout()
plt.show()
```

```python
# Bölümlere göre ortalama başarı
bolum_ortalama = df.groupby("Bölüm")[["Matematik", "Fizik", "Kimya"]].mean()

# Bar plot (her ders ayrı renkte)
bolum_ortalama.plot(kind="bar", figsize=(8,6))
plt.title("Bölümlere Göre Ders Ortalamaları")
plt.ylabel("Ortalama Not")
plt.xlabel("Bölüm")
plt.ylim(0, 100)
plt.xticks(rotation=0)  # Bölüm isimlerini yatay yapar
plt.legend(title="Ders")
plt.show()
```
