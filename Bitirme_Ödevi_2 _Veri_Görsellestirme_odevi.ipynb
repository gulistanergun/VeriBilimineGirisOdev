```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('50_Startups.csv')
```

```python
data.head()
```

## Bu veri seti, 50 farklı startup şirketinin çeşitli harcamalarını ve kârlılıklarını içermektedir.  

R&D Spend (Ar-Ge Harcaması): Şirketin araştırma ve geliştirme (Ar-Ge) için harcadığı tutar.  
Administration (Yönetim Harcaması): Şirketin yönetim giderleri için harcadığı tutar.  
Marketing Spend (Pazarlama Harcaması): Şirketin pazarlama ve reklam faaliyetleri için harcadığı tutar.
State (Eyalet): Şirketin faaliyet gösterdiği eyalet (örneğin, New York, California, Florida).  
Profit (Kâr): Şirketin elde ettiği toplam kâr.  
Bu veri seti, startup'ların çeşitli harcama kalemleri ile kârlılıkları arasındaki ilişkileri analiz   etmek için kullanılabilir. Örneğin, Ar-Ge veya pazarlama harcamalarının kârlılık üzerindeki etkisini   incelemek için uygun bir veri setidir.  

## 1.GÖREV : R&D Harcaması ve Kâr Arasındaki İlişki (Scatter Plot): Ar-Ge harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.

```python

plt.figure(figsize=(8, 5))
plt.scatter(data['R&D Spend'], data['Profit'], color='green', alpha=0.6)
plt.xlabel('R&D Harcaması', fontsize=12, )
plt.ylabel('Kâr', fontsize=12, )
plt.title('R&D Harcaması ve Kâr Arasındaki İlişki', fontsize=14, color='darkred',  pad=20)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Analiz: Grafik, Ar-Ge harcamalarının kârlılıkla güçlü bir pozitif korelasyon gösterdiğini ortaya koyuyor.
#Ar-Ge'ye daha fazla yatırım yapan şirketlerin genel olarak daha yüksek kâr elde ettiği görülmektedir.


```

## 2.GÖREV: Yönetim Harcamaları ve Kâr Arasındaki İlişki (Scatter Plot): Yönetim harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.

```python


plt.figure(figsize=(8, 5))
plt.scatter(data['Administration'], data['Profit'], color='green', alpha=0.6)
plt.xlabel('Yönetim Harcaması', fontsize=12)
plt.ylabel('Kâr', fontsize=12)
plt.title('Yönetim Harcaması ve Kâr Arasındaki İlişki', fontsize=14,  color='darkred', pad=20)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Analiz: Yönetim harcamalarının kâr üzerindeki etkisi Ar-Ge kadar belirgin değil.
# Dağılım daha heterojen bir yapı sergiliyor.

```

## 3. GÖREV: Eyaletlere Göre Ortalama Kâr (Bar Chart): Farklı eyaletlerdeki startup'ların ortalama kârlarını karşılaştıran bir çubuk grafik.

```python



plt.figure(figsize=(8, 5))
ortalama_kar = data.groupby('State')['Profit'].mean().sort_values(ascending=False)
plt.bar(ortalama_kar.index, ortalama_kar.values, color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
plt.xlabel('Eyalet', fontsize=12)
plt.ylabel('Ortalama Kâr', fontsize=12)
plt.title('Eyaletlere Göre Ortalama Kâr', fontsize=14,  color='darkred',pad=20)
plt.xticks(rotation=0)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# Analiz: Eyaletler arası kârlılık farklılıkları görülmektedir.
# Bu durum bölgesel pazar koşulları, vergi politikaları ve iş ortamının önemini vurgulamaktadır.

```

## 4. GÖREV: Harcama Türlerinin Karşılaştırması (Boxplot): R&D, yönetim ve pazarlama harcamalarının dağılımını karşılaştıran bir kutu grafiği.

```python

plt.figure(figsize=(10, 5))
harcamalar = [data['R&D Spend'], data['Administration'], data['Marketing Spend']]
labels = ['R&D Harcaması', 'Yönetim Harcaması', 'Pazarlama Harcaması']
plt.boxplot(harcamalar, tick_labels=labels, patch_artist=True)
plt.ylabel('Harcama Miktarı', fontsize=12)
plt.title('Harcama Türlerinin Karşılaştırması', fontsize=14, color='darkred', pad=20)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# Analiz: Pazarlama harcamalarının diğer kategorilere göre daha geniş bir dağılım gösterdiği,
# Ar-Ge harcamalarının ise daha homojen bir yapıya sahip olduğu görülmektedi


```
